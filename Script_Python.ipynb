```python
import requests
import pandas as pd

# Informations API
api_key = "adb1cd4172msh38b1ccc3d549a41p1b1bbcjsn1f6c6bf73118"
api_host = "edamam-food-and-grocery-database.p.rapidapi.com"
url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

# Paramètres de requête
params = {"ingr": "champagne"}

# En-têtes pour l'API
headers = {
    "x-rapidapi-key": api_key,
    "x-rapidapi-host": api_host
}

# Envoyer une requête GET à l'API
response = requests.get(url, headers=headers, params=params)

# Vérifier la réponse de l'API
if response.status_code == 200:
    data = response.json()

    # Récupérer les 10 premiers produits
    produits = data.get('hints', [])[:10]

    # Structurer les données en nettoyant les valeurs de foodContentsLabel
    product_list = []
    for item in produits:
        produit = item['food']
        # Nettoyer les valeurs (tabulations et espaces multiples remplacées par des virgules)
        contents_label = produit.get('foodContentsLabel', 'N/A').replace('\t', ',').replace('  ', ' ')
        product_list.append({
            'foodId': produit.get('foodId', 'N/A'),
            'label': produit.get('label', 'N/A'),
            'category': produit.get('category', 'N/A'),
            'foodContentsLabel': contents_label,
            'image': produit.get('image', 'N/A')
        })

    # Créer un DataFrame avec pandas pour organiser les données correctement
    df = pd.DataFrame(product_list)

    # Exporter vers un fichier CSV avec un séparateur de point-virgule
    df.to_csv('/content/drive/My Drive/projet_6/produits_champagne_clean_fixed.csv', sep=';', index=False)
    print("Fichier CSV créé avec succès.")
else:
    print(f"Erreur lors de l'accès à l'API : {response.status_code}")
```
